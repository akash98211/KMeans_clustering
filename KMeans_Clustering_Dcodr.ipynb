```python
#import standard Scaler
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
#Selecting the features to standardize
df_std = df.drop(['Id','Species'],axis=1)
```

```python
df_std.head()
```

```python
df_std = std.fit_transform(df_std)
```

```python
df_std
```

```python
df_std = pd.DataFrame(df_std)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df_std
```

```python
#import K Means Clustering
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3,random_state=1)
```

```python
kmeans.fit_transform(df_std)
```

```python
kmeans.labels_
```

```python
df.head()
```

```python
df['Cluster'] = kmeans.labels_
```

```python
df
```

```python
df['Cluster'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_plot = df.drop(['Id','Species'],axis=1)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.pairplot(df_plot,hue='Cluster')
plt.show()
```

```python
#Elbow Method
wss = []
cluster_list = list(range(1,6))
for x in cluster_list:
    kmeans = KMeans(n_clusters=x,init='k-means++',max_iter=300,n_init=10,random_state=42)
    kmeans.fit(df_std)
    wss.append(kmeans.inertia_)
```

```python
#Plot the elbow method
plt.plot(cluster_list,wss)
plt.title('Elbow_Method',fontsize=20)
plt.xlabel('No. of Clusters')
plt.ylabel('WSS')
plt.show()
```
